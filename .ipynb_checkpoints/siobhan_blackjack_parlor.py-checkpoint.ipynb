### Rules:

`1. ` The game will have two players: the Dealer and the Player. The game will start off with a deck of 52 cards. The 52 cards will consist of 4 different suits: Clubs, Diamonds, Hearts and Spades. For each suit, there will be cards numbered 1 through 13. <br>
**Note: No wildcards will be used in the program**

`2. ` When the game begins, the dealer will shuffle the deck of cards, making them randomized. After the dealer shuffles, it will deal the player 2 cards and will deal itself 2 cards from. The Player should be able to see both of their own cards, but should only be able to see one of the Dealer's cards.
 
`3. ` The objective of the game is for the Player to count their cards after they're dealt. If they're not satisfied with the number, they have the ability to 'Hit'. A hit allows the dealer to deal the Player one additional card. The Player can hit as many times as they'd like as long as they don't 'Bust'. A bust is when the Player is dealt cards that total more than 21.

`4. ` If the dealer deals the Player cards equal to 21 on the **first** deal, the Player wins. This is referred to as Blackjack. Blackjack is **NOT** the same as getting cards that equal up to 21 after the first deal. Blackjack can only be attained on the first deal.

`5. ` The Player will never see the Dealer's hand until the Player chooses to 'stand'. A Stand is when the player tells the dealer to not deal it anymore cards. Once the player chooses to Stand, the Player and the Dealer will compare their hands. Whoever has the higher number wins. Keep in mind that the Dealer can also bust. 

```python
import pprint   # PrettyPrinter.pprint(object)   # 
import random   # random.shuffle(x[, random])
import time     # time.sleep(secs)
```

```python
class Card:
    def __init__(self, card_tuple):
        self.suit = card_tuple[0]
        self.order = card_tuple[1]

    # print name of card
    def __str__(self):
        return f"{self.order} of {self.suit}"

    # assign point value to card
    def get_points(self):
        if self.order.isdigit():
            return int(self.order)
        elif self.order == 'Ace':
            return 11
        else:
            return 10

    def hidden(self):
        return '| HIDDEN |'
```

```python
class Player:

    def __init__(self, name):
        self.name = name.title()
        self.hand = []

    def __str__(self):
        return f"{self.name}"
    
    def __repr__(self):
        return f"< Player name: '{self.name}' >\n< Player's hand: {self.display_hand()} >"
    
    # count points in hand
    # account for Ace having 2 different values
    def hand_total(self):
        total = 0
        for card in self.hand:
            total += card.get_points()
            if card.order == 'Ace' and total > 21:
                total -= 10
        return total

    def hit(self):
        return self.hand_total() < 21

    def win(self):
        return self.hand_total() == 21
        
    def bust(self):
        return self.hand_total() > 21
    
    # set method to tell whether player gets blackjack on first turn    
    def blackjack(self):
        return len(self.hand) == 2 and self.win()
    
    def reveal(self, show = True):
        if show == False:
            return f"| HIDDEN | "
        else:
            return f"{self.hand[0].order} of {self.hand[0].suit}, "     
    
    def display_hand(self, show):
        cards = ''
        if show == False:
            cards = self.reveal(False)
        else:
            cards = self.reveal(True)
        for card in self.hand[1:-1]:
            cards += f'{card.order} of {card.suit}, '
        cards += f'{self.hand[-1].order} of {self.hand[-1].suit}'
        if show == False:
            return f"{self.name.title()}'s hand: {cards}"
        else:
            return f"{self.name.title()}'s hand: {cards} ==> {self.hand_total()}"
    
class Dealer(Player):
    def __init__(self, name = 'Dealer'):
        super().__init__(name)
        
    # tests whether dealer should stand-   
    def stand(self):
        return self.hand_total() >= 17
            
class Gambler(Player):
    def __init__(self, name):
        super().__init__(name)
        if self.name == '':
            self.name = 'Gambler'
        self.pot = 100
        
    def calc_bet(self, wager, outcome):
        if outcome == 'won':
            self.pot += wager
        elif outcome == 'lost':
            self.pot -= wager    
        else:
            print(f"Could not calculate {self}'s winnings")
        return self.pot
```

```python
class Blackjack:
    
    def __init__(self, dealer, player, bet = None):
        self.dealer = dealer
        self.player = player
        self.bet = bet
        self.deck = []
        self.num_games = 0
        self.wins = 0
        self.losses = 0
        
    def place_bet(self):
        bet_ask = input ("Would you like to place a bet? (y/n)  ").lower()
        while bet_ask not in {'y', 'n'}:
            bet_ask = input ("Please choose 'Y' or 'N'  ").lower()
        if bet_ask == 'y':
            print(f"{self.player} currently has ${self.player.pot}")
            self.bet = int(input("\tHow much would you like to wager? "))
            while self.bet > self.player.pot:
                print(f"Sorry, {self.player}, you do not have enough to place that bet.")
                self.bet = int(input(f"\tEnter a number less than {self.player.pot}: "))
        else:
            self.bet = 0
            
#     def double(self):
#         if len(self.deck) == 2 and 8 > self.player.hand_total() > 12:
#             double_ask = input("Would you like to double? (y/n)  ").lower()
#             while double_ask not in {'y', 'n'}:
#                 double_ask = input("Double? Please type 'Y' or 'N'  ")
#             if double_ask == 'y':
#                 self.bet *= 2
#                 return True
#             else:
#                 return False
    
    def get_deck(self):
        suits = ['Hearts', 'Spades', 'Diamond', 'Clubs']
        orders = ['Ace', '2', '3', '4', '5', '6', '7', '8', '9', '10', 'Jack', 'Queen', 'King']
        self.deck = [(suit, order) for suit in suits for order in orders]
        random.shuffle(self.deck)
        return self.deck   
        
    def get_card(self, person):
        card = Card(self.deck.pop())
        person.hand.append(card)
        return card   
    
    def setup(self):
        self.dealer.hand = []
        self.player.hand = []
        deck = self.get_deck()
        print("\nShuffling deck...")
        time.sleep(2)
        print("Dealing cards...\n")
        time.sleep(1)
        for i in range(0,2):
            self.get_card(self.dealer)
            self.get_card(self.player)
        print(f"\n{self.dealer.display_hand(False)}")
        print(f"{self.player.display_hand(True)}\n")
#         if self.bet != None:
#             if self.double() == True:
#                 self.get_card(self.player)
#                 print(f"\n{self.dealer.display_hand(False)}")
#                 print(f"{self.player.display_hand(True)} and \n")
        if self.player.blackjack():
            print(f"\n{self.player} gets BLACKJACK!\n\nYou win :)")
            self.end()
    
    def dealer_turn(self):
        print("\nDealer's turn...\n")
        time.sleep(1)
        while self.dealer.stand() == False:
            self.get_card(self.dealer)
            
    def go_on(self, turn):
        if turn == 'player':
            self.get_card(self.player)
            show = False
        else:
            self.dealer_turn()
            show = True
        print(f"\n{self.dealer.display_hand(show)}")
        print(f"{self.player.display_hand(True)}\n")
    
    def record(self, outcome):
        self.num_games += 1
        if outcome == 'won':
            self.wins += 1
        else:
            self.losses += 1
        print(f"Games played: {self.num_games}\nGames won: {self.wins}\nGames lost: {self.losses}")
            
    def end(self):
        outcome = ''
        if self.dealer.win() and self.player.win() == False:
            print(f"\nDealer hit *** 21 ***\n{self.player} loses :(")
            outcome = 'lost'
        elif self.player.win() and self.player.blackjack() == False:
            print(f"\n{self.player} hit *** 21 *** \nYou win :)")
            outcome = 'won'
        elif self.dealer.bust() and self.player.hit():
            print(f"\nDealer BUSTS. {self.player} wins :)\n")
            outcome = 'won'
        elif self.player.bust() and self.dealer.hit():
            print(f"\n{self.player} BUSTS :( Dealer wins.\n")
            outcome = 'lost'
        else:
            winner = self.player
            outcome = 'won'
            if self.player.hand_total() == self.dealer.hand_total():
                print("\n* * * Wowee! It's a tie! What a lucky game :) * * * \n")
            else: 
                self.player.hand_total() <  self.dealer.hand_total()
                winner = self.dealer
                outcome = 'lost' 
            print(f"{winner} wins!\n")
        self.record(outcome)
        print(f"\nFinal Score =\n{self.player}: {self.player.hand_total()}\n{self.dealer}: {self.dealer.hand_total()}")
        if self.bet != None:
            pot = self.player.calc_bet(self.bet, outcome)
            print(f"\t{self.player} {outcome} {self.bet} and now has ${self.player.pot}\n")
        
        
```
